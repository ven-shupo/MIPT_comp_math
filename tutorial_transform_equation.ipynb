# Convection equation
$$\frac{dU(x,t)}{dt} + a*\frac{dU(x,t)}{dx} = f$$.
$$U(x,0) = U_1(x)$$
$$U(0,t) = U_2(t)$$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import *
```

```python
from transfom_equation import convection_equation
```

```python
# set the initial, boundary conditions and function f
def U1(x):
    """
    set the initial condition
    
    int: the point
    out: value of func in this point
    """
    return sin(4*pi*x/20)

def U2(t):
    """
    set the border condition
    
    int: the point
    out: value of func in this point
    """
    return 0

def f(x, t):
    """ 
    set the right func
    
    int: x,t - the point x and t
    out: value of func in this points
    """
    return 0
```

```python
A = convection_equation()
A.fit(1, 18, 20, 360, 400, U1 = U1, U2 = U2, frequency = False)
U_1 = A.get_U()
X1 = A.get_X()
T1 = A.get_T()
```

```python
B = convection_equation()
B.fit(1, 18, 20, 3600, 400, U1 = U1, U2 = U2, f= f, frequency = True)
U_2 = B.get_U()
X2 = B.get_X()
T2 = B.get_T()
```

```python
# moment in time which we draw a graph
t1 = [0, int(len(T1)/4), int(len(T1)/2), int(3*len(T1)/4),-1]
t2 = [0, int(len(T2)/4), int(len(T2)/2), int(3*len(T2)/4),-1]
time = [0,4,9,12,18]

fig, axs = plt.subplots(1,5,figsize=(20,4))

# draw the graph
for i in range(1,6):
    axs[i-1].plot(X1,U_1[t1[-i]])
#     axs[i-1].plot(X2,U_2[t2[-i]],'--')
    axs[i-1].set_title('U(X) in t ='f"{time[i-1]}")
    axs[i-1].grid()
    axs[i-1].axis([0,20,-1.1,1.1])

axs[0].legend([f"{A.method, A.frequency}",f"{B.method, B.frequency}"])
    
None
```
